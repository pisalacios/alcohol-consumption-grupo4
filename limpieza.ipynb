```python
import pandas as pd
import openpyxl

csv = ('https://raw.githubusercontent.com/pisalacios/alcohol-consumption-grupo4/refs/heads/main/15.%20Alcohol%20consumption.csv')

df = pd.read_csv(csv)

    # Eliminar columnas innesesarias y renombrar las de interés

df = df.drop(columns=['sex','IndicatorCode','iso3'])
df = df.rename(columns={'country':'pais',
                        'alcohol_liters_per_capita':'litros_por_persona',
                        'year':'año',
                        'lower_ci':'limite_inf',
                        'upper_ci':'limite_sup',
                        'ci_width':'rango'}
)

    # Filtrando por paises y años de interés

paises_europeos = [
'Albania', 'Andorra', 'Austria', 'Belarus', 'Belgium', 'Bosnia and Herzegovina', 
    'Bulgaria', 'Croatia', 'Cyprus', 'Czechia', 'Denmark', 'Estonia', 'Finland', 
    'France', 'Germany', 'Greece', 'Hungary', 'Iceland', 'Ireland', 'Italy',
    'Latvia', 'Lithuania', 'Luxembourg', 'Malta', 'Moldova, Republic of', 
    'Montenegro', 'Netherlands', 'North Macedonia', 'Norway', 'Poland', 
    'Portugal', 'Romania', 'Russian Federation', 'Serbia', 'Slovakia', 
    'Slovenia', 'Spain', 'Sweden', 'Switzerland', 'Ukraine', 'United Kingdom'
] 

años_interes =  [ 2016, 2017, 2018,2019, 2020, 2022]
df_filtrado = df[(df['pais'].isin(paises_europeos)) & (df['año'].isin(años_interes))]

    # Redondeando a dos decimales 

columnas_a_redondear = ['litros_por_persona', 'limite_inf', 'limite_sup', 'rango']
df_filtrado[columnas_a_redondear] = df_filtrado[columnas_a_redondear].round(2)

df_filtrado = df_filtrado.sort_values(by=['pais', 'año'])

    # Añadiendo nueva clausula de temporada de interés

def clasificar_temporada(año):
    if año <= 2019:
        return "Antes"
    elif año == 2020:
        return "Durante"
    else:
        return "Después"

df_filtrado['temporada'] = df_filtrado['año'].apply(clasificar_temporada)

df_filtrado
```

```python

    # Exportar el DF filtrado a un nuevo Excel

df_filtrado.to_excel('alcohol_filtrado.xlsx', index=False)

print(df_filtrado['año'].value_counts())
```
