```python
import pandas as pd
import openpyxl as op

csv = ('https://raw.githubusercontent.com/pisalacios/alcohol-consumption-grupo4/refs/heads/main/15.%20Alcohol%20consumption.csv')
df = pd.read_csv(csv)

    # Eliminar columnas innesesarias y renombrar las de interés

df = df.drop(columns=['sex','IndicatorCode','iso3'])
df = df.rename(columns={'country':'pais',
                        'alcohol_liters_per_capita':'litros_por_persona',
                        'year':'año',
                        'lower_ci':'limite_inf',
                        'upper_ci':'limite_sup',
                        'ci_width':'rango'}
)

    # Traduciendo los países de interés

paises_traducidos = {
    'Azerbaijan': 'Azerbaiyán',
    'Belarus': 'Bielorrusia',
    'Belgium': 'Bélgica',
    'Bosnia and Herzegovina': 'Bosnia y Herzegovina',
    'Croatia': 'Croacia',
    'Cyprus': 'Chipre',
    'Czechia': 'República Checa',
    'Denmark': 'Dinamarca',
    'Finland': 'Finlandia',
    'France': 'Francia',
    'Germany': 'Alemania',
    'Greece': 'Grecia',
    'Hungary': 'Hungría',
    'Iceland': 'Islandia',
    'Ireland': 'Irlanda',
    'Italy': 'Italia',
    'Kazakhstan': 'Kazajistán',
    'Latvia': 'Letonia',
    'Lithuania': 'Lituania',
    'Luxembourg': 'Luxemburgo',
    'Netherlands': 'Países Bajos',
    'North Macedonia': 'Macedonia del Norte',
    'Norway': 'Noruega',
    'Poland': 'Polonia',
    'Republic of Moldova': 'República de Moldavia',
    'Romania': 'Rumanía',
    'Russian Federation': 'Federación de Rusia',
    'Slovakia': 'Eslovaquia',
    'Slovenia': 'Eslovenia',
    'Spain': 'España',
    'Sweden': 'Suecia',
    'Switzerland': 'Suiza',
    'Türkiye': 'Turquía',
    'Ukraine': 'Ucrania',
    'United Kingdom': 'Reino Unido'
}

    # Filtrando países y años de interés 

paises_europeos = [
    'Albania', 'Andorra', 'Armenia', 'Austria', 'Azerbaijan', 'Belarus', 'Belgium', 
    'Bosnia and Herzegovina', 'Bulgaria', 'Croatia', 'Cyprus', 'Czechia', 'Denmark', 
    'Estonia', 'Finland', 'France', 'Georgia', 'Germany', 'Greece', 'Hungary', 
    'Iceland', 'Ireland', 'Italy', 'Kazakhstan', 'Latvia', 'Lithuania', 'Luxembourg', 
    'Malta', 'Montenegro', 'Netherlands', 'North Macedonia', 'Norway', 'Poland', 
    'Portugal', 'Republic of Moldova', 'Romania', 'Russian Federation', 'San Marino', 
    'Serbia', 'Slovakia', 'Slovenia', 'Spain', 'Sweden', 'Switzerland', 'Türkiye', 
    'Ukraine', 'United Kingdom'
]

años_interes =  [ 2016, 2017, 2018, 2019, 2020, 2021, 2022]

df_filtrado = df[(df['pais'].isin(paises_europeos)) & (df['año'].isin(años_interes))]

    # Aplicando traducción

df_filtrado['pais'] = df_filtrado['pais'].replace(paises_traducidos)

    # Redondeando a dos decimales 

columnas_a_redondear = ['litros_por_persona', 'limite_inf', 'limite_sup', 'rango']
df_filtrado[columnas_a_redondear] = df_filtrado[columnas_a_redondear].round(2)

df_filtrado = df_filtrado.sort_values(by=['pais', 'año'])

    # Añadiendo nueva clausula de temporada de interés

def clasificar_temporada(año):
    if año <= 2019:
        return "Antes"
    if año == 2021:
        return 'No aplica'
    elif año == 2020:
        return "Durante"
    else:
        return "Después"

df_filtrado['temporada'] = df_filtrado['año'].apply(clasificar_temporada)

df_filtrado
```

```python
import os

    # Definiendo ruta y carpeta de manera compacta
ruta_carpeta = 'Data/Clean'
nombre_archivo = 'Alcohol_Filtrado.xlsx'
ruta_completa = os.path.join(ruta_carpeta, nombre_archivo)

os.makedirs(ruta_carpeta, exist_ok=True)

    # Exportar el DF filtrado a un nuevo Excel
df_filtrado.to_excel(ruta_completa, index=False)
```
